```python
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')

# This function converts a string into list
```

```python
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L

# Defining a function to convert string into list 
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
def convert4(obj):
    L=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter!=3:
         L.append(i['name'])
         counter+=1
        else:
           break
    return L

# Through this we are only fetching the name of first 3 cast
```

```python
movies['cast']=movies['cast'].apply(convert4)
```

```python
movies['cast'][0]
```

```python
movies['crew'][0]
```

```python
def fetch_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
            break
    return L

# We have to only fetch the name of the director from the crew
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","")for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","")for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","")for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","")for i in x])

# We have used this coode to remove spaces between each values
```

```python
movies['tags']=movies['overview']+movies['keywords']+ movies['cast']+movies['crew']+movies['genres']

# Here we have created new field called tags which ii a concatenation of above 5 fields 
```

```python
new_df= movies[['movie_id','title','tags']]

# Here we have created a new dataframe called new_df 
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x))

# As tags field in list formate we have converted it in to strings separated by spaces
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())

# We converting everything to lowercase as its recommended mostly
```

```python
new_df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')

# We imported this to perform text vectorisation
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()

''' fit_transform -- Learns all the unique words in the "tags" column.
For example, from the tags above, it might learn: ["space", "future", "aliens", "pirates", "ocean", "adventure", "spy", "mission", "action"]
Creates a numerical representation for each movie:
For each tag, it counts how many times each word appears 
And then converts it into 2d array'''
```

```python
vectors[0]
```

```python
cv.get_feature_names_out()
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()


""" 
We have imported this lib so that we can remove similar words like (action,actions, etc)
The Porter Stemmer is a common algorithm used in text processing to reduce words to their base or root form.
For example:
"running" â†’ "run"
"going" â†’ "go" 

"""
```

```python
def stem(text):
    y=[]
    for i in text.split():
      y.append(ps.stem(i))
    return " ".join(y)

''' 
The function stem you've written is designed to take a text (a string), split it into individual words, apply 
stemming to each word using the PorterStemmer, and then join the stemmed words back together into a string.

Breakdown of the stem function:
Input: A string of text (e.g., a sentence or paragraph).
Process:
Split the text into individual words using the .split() method.
Apply the stemming process to each word using the ps.stem(i) method.
Output: A string where the words have been stemmed, and they are rejoined into a sentence using " ".join(y).
'''
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

#This function is commonly used to measure the similarity between two vectors 
# or between all pairs of vectors in a dataset.
```

```python
similarity=cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
list(enumerate(similarity[0]))

# This fn converts similarity list to tuples so that the index of every movies is attached with it and does no
#  get replaced while sorting
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]

# The key in above code is used to tell that we are performing sorting on the basises of 1st index and not 0th index
# [1:6] this is used for only diplaying top 5 similar movies
```

```python
def recommend(movie):
    movie_index=new_df[new_df['title']==movie].index[0]
    distances=similarity[movie_index]
    movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]

    for i in movies_list:
        print(new_df.iloc[i[0]].title)

   
```

```python
recommend('Batman Begins')
```
